# Herramienta GIT

## Sistema de control de versiones


Un sistema de control de versiones es un software que le permite al usuario contar con un orden sobre los archivos desarrollados para un proyecto. Estos sistemas facilitan el versionamiento de codigos, ya que al momento de actualizar, eliminar o crear archivos se "guarda" una version con los cambios nuevos y los anteriores creando una especie de arbol de los distintos archivos.

La forma en que Git maneja la informaci√≥n almacena es a traves de una serie de archivos y las modificaciones hechas al archivo se guardan "debajo" del mismo, de manera que retiene el archivo anterior y toma una "imagen" de este, Git maneja los archivos con una secuencia, si el archivo es igual que el anterior, solamente hace un enlace al archivo anterior.

Git no permite cambiar los archivos sin que el sistema lo pase por alto ya que hace un checksum por cada archivo, esto quiere decir que los archivos no pueden ser corrompidos.


Git necesita pocos recursos de la pc, ya que lee directamente de la base de datos local, por lo que tampoco se estar conectado a internet.

#### Estados en GIT

Los estados de GIT pueden ser 3:
<ol>
<li>committed</li>
<li>modified</li>
<li>staged</li>
</ol>

<strong> committed </strong> que significa que el archivo esta almacenado. <br>
<strong> modified </strong> que significa que el archivo esta modificado pero no almacenado.<br>
<strong> Staged </strong> que significa que se marco un archivo modificado en la actual version, marcada para el proximo modified.<br>
